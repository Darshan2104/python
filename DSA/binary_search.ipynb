Search Single Element in a sorted array [25 min]
1. greedy
2. Xor
3. Binary search

```python
%%time
def find_single_element(arr):
    left = 0
    right = len(arr)-1
    element = -1
    mid = (left + right)//2

    while left <= right:
        mid = (left + right)//2
        # print("\nleft : ",left)
        # print("mid : ",mid)
        # print("right : ",right,"\n")
        if (left == right) or ((mid+1<right and mid-1>left) and (arr[mid-1]!= arr[mid] and arr[mid]!= arr[mid+1])):
            element = [arr[mid]]
            break
        elif arr[mid] == arr[mid+1]:
            if (abs(right-mid)+1)&1:
                left = mid
                # odd
            else:
                # even
                right = mid -1
        else:
            if (abs(mid-left)+1)&1:
                right = mid
                # odd
            else:
                # even
                left = mid+1
            


    return element
    

input_arr = [1,2,2,3,3,8,8]
print(find_single_element(input_arr))
```

Problem : Search element in a sorted and rotated array/ find pivot where it is rotated [20 min]
1. Linear Search
2. Binary Search

```python
def find_element(arr,target):
    left = 0
    right = len(arr)-1

    if arr[right] < target < arr[left]:
            return -1
    while left <= right:
        mid = ((left+right)//2)
        # print("\nleft :",left)
        # print("mid :",mid)
        # print("right :",right,"\n")
        if arr[left] == target:
            return left
        if arr[right] == target:
            return right
        if arr[mid] == target:
            return mid
        
        if (arr[mid]>arr[left]):
            if arr[left] < target < arr[mid]:
                right = mid-1
            else:
                left = mid + 1
        else:
            if arr[mid] < target < arr[right]:
                left = mid+1
            else:
                right = mid -1
       
    return -1

arr = [2,3,4,5,0,1]
target = 0
find_element(arr,target)
```

### 1. Fibonacci [LeetCode](https://leetcode.com/problems/fibonacci-number/description/)

````python
class Solution:
    def fib(self, n: int) -> int:
        prev = 0
        curr = 1
        if n <=1:
            return n
        else:
            for i in range(n-1):
                temp = curr
                curr = curr + prev
                prev = temp
            
            return curr
````

### 2. Climbing Stairs [LeetCode](https://leetcode.com/problems/climbing-stairs/description/)

````python
class Solution:
    def climbStairs(self, n: int) -> int:
        '''
        # Recursion will give TLE
        # Memorization ::>
        
        memo = [-1]*(n+1)
        def f(index):
            if index == 0 or index == 1:
                return 1
            if memo[index] != -1:
                return memo[index]
            one_step = f(index-1)
            two_step = f(index-2) if index > 1 else 0
            memo[index] = one_step + two_step
            return memo[index]
        return f(n)
        '''

        # Tabulation method ::>
        memo = [1]
        
        for i in range(1,n):
            
            one_step = memo[i-1]
            two_step = memo[i-2] if i > 1 else 1
            memo.append(one_step + two_step) 
        return memo[-1]
````

### 3. Frog Jump [GFG](https://www.geeksforgeeks.org/problems/geek-jump/1)

- What if frog can do upto k jumps?
- ANS : run sub for loop from 1->k

````python
#User function Template for python3
class Solution:
    def minCost(self, height):
        
        '''
        # Memorization 
        
        n = len(height)
        memo = [-1]*(n+1)
        def f(i):
            if i == 0:
                return 0
            if memo[i] != -1:
                return memo[i]
            one_step = f(i-1) + abs(height[i]-height[i-1]) 
            two_step =f(i-2) + abs(height[i]-height[i-2]) if i > 1 else float('inf')
            memo[i] = min(one_step, two_step)
            return memo[i]
        return f(len(height)-1)
        '''
        
        # Tabulation (Bottom-up))
        n = len(height)
        
        memo = [0]
        
        for i in range(1,n):
            one_step = memo[i-1] + abs(height[i] - height[i-1])
            two_step = memo[i-2] + abs(height[i] - height[i-2]) if i > 1 else float('inf')
            
            memo.insert(i,min(one_step,two_step))
        
        return memo[-1]
        
````

### 6. Unique Paths [LeetCode](https://leetcode.com/problems/unique-paths/description/)

````python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:

        # All possible values ---> Recursion ---> DP ---> Memorization ---> Bottom-up Tabulization


        # Memorization
        '''
        memo = [[-1]*n for _ in range(m)]
        def f(i,j):
            if i == 0 and j == 0:
                return 1
            if i < 0 or j < 0:
                return 0
            if memo[i][j] != -1:
                return memo[i][j]

            up  = f(i-1,j)
            left = f(i, j-1)
            memo[i][j] = up + left
            return memo[i][j]
        
        return f(m-1, n-1)
        '''

        #  Tabulation : Bottom-up
        memo = [[-1]*n for _ in range(m)]
        
        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    memo[0][0] = 1
                else:
                    up = memo[i-1][j] if i>0 else 0
                    left = memo[i][j-1] if j>0 else 0
                    memo[i][j] = left + up
        return memo[-1][-1]
                
````

### 8. Minimum Path Sum [LeetCode](https://leetcode.com/problems/minimum-path-sum/)

````python
class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        # minimization ---> DP
        '''
        # memoization
        m,n = len(grid), len(grid[0])
        memo = [ [-1]*(n) for _ in range(m)]
        def f(i,j):
            # Termination condition
            if i < 0 and j < 0:
                return float('inf')
            if i == 0 and j == 0:
                return grid[i][j]
            
            if memo[i][j] != -1:
                return memo[i][j]
            
            # All the things we can doo
            up = grid[i][j] + f(i-1,j) if i>0 else float('inf')
            left = grid[i][j] + f(i,j-1) if j>0 else float('inf')

            # Pick the best (minimization)
            memo[i][j] = min(up, left)
            return memo[i][j]     

        return f(m-1,n-1)
        '''
        m,n = len(grid), len(grid[0])
        memo = [ [-1]*(n) for _ in range(m)]

        for i in range(m):
            for j in range(n):
                if i == 0 and j==0:
                    memo[i][j] = grid[i][j]
                else:
                    up = grid[i][j] + memo[i-1][j] if i>0 else float('inf')
                    left = grid[i][j] + memo[i][j-1] if j>0 else float('inf')
                    memo[i][j] = min(up, left)

        return memo[-1][-1]
````

### 9. Minimum Path Sum(Triangle) [LeetCode](https://leetcode.com/problems/triangle/description/)

````python
class Solution:
    def minimumTotal(self, triangle: List[List[int]]) -> int:
        
        '''
        # Memorization
        n = len(triangle)
        memo = [[-1]*n for _ in range(n)]
        def f(i,j):
            if i== (n-1):
                return triangle[i][j]
            
            if memo[i][j] != -1:
                return memo[i][j]
            down = triangle[i][j] + f(i+1,j)
            diag = triangle[i][j] + f(i+1,j+1)
            memo[i][j] =  min(down, diag)
            return memo[i][j]
        
        return f(0,0)
        '''
        n = len(triangle)
        memo = [[-1]*n for _ in range(n)]

        for j in range(n):
            memo[n-1][j] = triangle[n-1][j]
        
        for i in range(n-2,-1,-1):
            for j in range(i,-1,-1):
                down = triangle[i][j] + memo[i+1][j]
                diag = triangle[i][j] + memo[i+1][j+1]
                memo[i][j] =  min(down, diag)

        return memo[0][0]
````

### 10. Minimum/Maximum Falling Path Sum [Leetcode](https://leetcode.com/problems/minimum-falling-path-sum/description/)

### 11. cherry-pickup-II [Leetcode](https://leetcode.com/problems/cherry-pickup-ii/description/)

````python
class Solution:
    def cherryPickup(self, grid: List[List[int]]) -> int:
        m,n = len(grid), len(grid[0])

        l = max(m,n)
        memo = [[[-1]*l for _ in range(l)] for _ in range(l)]
        
        '''
        # Memorization
        def f(i,j1, j2):
            if (j1 < 0 or j1 >= n) or (j2 < 0 or j2 >= n):
                return float('-inf')
            
            if i == m-1:
                return grid[i][j1] if j1 == j2 else grid[i][j1] + grid[i][j2]
            
            if memo[i][j1][j2] != -1:
                return memo[i][j1][j2]
            
            # Explore all possible values....
            maxi = 0
            for dj1 in [-1,0,1]:
                for dj2 in [-1,0,1]:
                    if j1 == j2:
                        maxi = max(maxi, grid[i][j1] + f(i+1, j1+ dj1, j2+dj2))
                    else:
                        maxi = max(maxi, grid[i][j1] + grid[i][j2] + f(i+1, j1+ dj1, j2+dj2))
            memo[i][j1][j2]= maxi
            return maxi
        return f(0,0,n-1)
        '''
        m, n = len(grid), len(grid[0])
        # Initialize 3D DP table with -1
        dp = [[[-1 for _ in range(n)] for _ in range(n)] for _ in range(m)]
        
        # Base case: fill for the last row
        for j1 in range(n):
            for j2 in range(n):
                if j1 == j2:
                    dp[m-1][j1][j2] = grid[m-1][j1]
                else:
                    dp[m-1][j1][j2] = grid[m-1][j1] + grid[m-1][j2]
        
        # Bottom-up DP
        for i in range(m-2, -1, -1):
            for j1 in range(n):
                for j2 in range(n):
                    maxi = float('-inf')
                    for dj1 in [-1, 0, 1]:
                        for dj2 in [-1, 0, 1]:
                            if 0 <= j1 + dj1 < n and 0 <= j2 + dj2 < n:
                                curr = grid[i][j1] if j1 == j2 else grid[i][j1] + grid[i][j2]
                                curr += dp[i+1][j1 + dj1][j2 + dj2]
                                maxi = max(maxi, curr)
                    dp[i][j1][j2] = maxi
        
        return dp[0][0][n-1]
````

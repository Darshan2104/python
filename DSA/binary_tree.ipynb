## 1.InOrder, PreOrder, PostOrder

```python
class Node:
    def __init__(self,key):
        self.value = key
        self.left = None
        self.right = None

def inorder_traversal(root):
    # Left - root - Right
    if root:
        inorder_traversal(root.left)
        print(root.value, end= " ")
        inorder_traversal(root.right)
    
def pre_order_traversal(root):
    # root - left - right
    if root:
        print(root.value, end=" ")
        pre_order_traversal(root.left)
        pre_order_traversal(root.right)

def post_order_traversal(root):
    # left - right - root
    if root:
        post_order_traversal(root.left)
        post_order_traversal(root.right)
        print(root.value,end=" ")

# if __name__ == "__main__()":
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

print("In-order")
inorder_traversal(root)
print("\nPre-order")
pre_order_traversal(root)
print("\nPost-order")
post_order_traversal(root)


root = Node(1)
root.left = Node(2)
root.left.left = Node(4)
root.left.right = Node(5)

root.left.left.left = Node(8)
root.left.left.right = Node(9)

root.left.right.left = Node(10)
root.left.right.right = Node(11)

root.right = Node(3)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.left.right = Node(13)
root.right.right.left = Node(14)

print("\n======================================")
inorder_traversal(root)
```

## BFS

```python
def bfs(root):
    queue = [root]
    ans = [root.value]
    while len(queue) > 0:
        root = queue.pop(0)
        if root.left is not None:
            queue.append(root.left)
            ans.append(root.left.value)
        if root.right is not None:
            queue.append(root.right)
            ans.append(root.right.value)
        
    return ans

    
bfs(root)
```

## Bottom View of a Binary Tree

```python
class  Node:
    def __init__(self, key):
        self.value = key
        self.horiontal_distance = float('inf')
        self.left = None
        self.right = None

def bottom_view_of_BT(root):
    bottom_view = []
    if root == None:
        return []
    
    hd_dict = {}
    q = []
    hd = 0
    root.horiontal_distance = hd
    
    q.append(root)
    hd_dict[hd] = root.value
    while q:    
        root = q.pop(0)
        hd = root.horiontal_distance
        if root.left != None:
            root.left.horiontal_distance = hd -1
            hd_dict[root.left.horiontal_distance] = root.left.value
            q.append(root.left)
            # store it in dictionay
        if root.right != None:
            root.right.horiontal_distance = hd +1
            hd_dict[root.right.horiontal_distance] = root.right.value
            q.append(root.right)

    for key in sorted(hd_dict.keys()):
        bottom_view.append(hd_dict[key])
        print(f"{key}--->{hd_dict[key]}")
    
    return bottom_view
    

    

# +++++++++++++++++++++++++++++++++++++++++++++++++
if __name__ == '__main__':
    # TREE 
    root = Node(20)
    root.left = Node(8)
    root.right = Node(22)
    root.left.left = Node(5)
    root.left.right = Node(3)
    root.right.left = Node(4)
    root.right.right = Node(25)
    root.left.right.left = Node(10)
    root.left.right.right = Node(14)
    # +++++++++++++++++++++++++++++++++++++++++++++++++

    ans = bottom_view_of_BT(root)
    print(ans)
```

## Vertical Order Traversal of Binary Tree

```python
class  Node:
    def __init__(self, key):
        self.value = key
        self.horiontal_distance = float('inf')
        self.left = None
        self.right = None

def vertical_order_traversal(root):
        ans = []
        if root == None: return ans

        hd_dict = {}
        q = []
        hd = 0
        root.horiontal_distance = hd
        
        q.append(root)
        hd_dict[hd] = [root.value]
        while q:    
            root = q.pop(0)
            hd = root.horiontal_distance
            if root.left != None:
                root.left.horiontal_distance = hd -1
                try:
                    hd_dict[root.left.horiontal_distance].append(root.left.value)
                except:
                    hd_dict[root.left.horiontal_distance] = [root.left.value]
                
                q.append(root.left)
                # store it in dictionay
            if root.right != None:
                root.right.horiontal_distance = hd +1
                try:
                    hd_dict[root.right.horiontal_distance].append(root.right.value)
                except:
                    hd_dict[root.right.horiontal_distance] = [root.right.value]
                
                q.append(root.right)
        for key in sorted(hd_dict.keys()):
            ans.append(hd_dict[key])
            print(f"{key}--->{hd_dict[key]}")
        return ans

if __name__ == '__main__':
    # TREE 
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    root.right.left.right = Node(8)
    root.right.right.right = Node(9)
    # root = Node(20)
    # root.left = Node(8)
    # root.right = Node(22)
    # root.left.left = Node(5)
    # root.left.right = Node(3)
    # root.right.left = Node(4)
    # root.right.right = Node(25)
    # root.left.right.left = Node(10)
    # root.left.right.right = Node(14)
# +++++++++++++++++++++++++++++++++++++++++++++++++

    ans = vertical_order_traversal(root)
    for i in ans:
        print(i)
```

## Maximum width of a Binary Tree

```python
class  Node:
    def __init__(self, key):
        self.value = key
        self.left = None
        self.right = None

def get_maximum_width(root):
        max_width = 1        
        if root.left == None and root.right == None: return max_width
        
        q = []
        q.append(root)
        while len(q) != 0:
            l = q
            max_width = max(max_width,len(l))
            q = []
            for i in l:
                if i.left != None:
                    q.append(i.left)
                if i.right != None:
                    q.append(i.right)
        return max_width



        

if __name__ == '__main__':
    # TREE 
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    root.right.left.right = Node(8)
    root.right.right.right = Node(9)
    # +++++++++++++++++++++++++++++++++++++++++++++++++
    print(get_maximum_width(root))
```

## Diameter of a Binary Tree

```python
class Node:
    def __init__(self, key):
        self.value = key
        self.left = None
        self.right = None

def hight_of_tree(root):
    if root is None:
        return 0
    else:
        l_hight = hight_of_BT(root.left)
        r_hight = hight_of_BT(root.right)

        return max(l_hight,r_hight) + 1

def diameter_of_BT(root):
    if root is None:
        return 0
    else:
        l_hight = hight_of_BT(root.left)
        r_hight = hight_of_BT(root.right)

        l_diameter = diameter_of_BT(root.left)
        r_diameter = diameter_of_BT(root.right)

        return max(l_hight + r_hight +1, max(l_diameter,r_diameter))

if __name__ == '__main__':
    # TREE 
    ####################################
    # root = Node(1)
    # root.left = Node(2)
    # root.right = Node(3)
    # root.left.left = Node(4)
    # root.left.right = Node(5)
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    root.right.left.right = Node(8)
    root.right.right.right = Node(9)
    root.right.right.right.right = Node(10)

    ####################################
    print(diameter_of_BT(root))
```

## ZigZag Tree Traversal


```python
class Node:
    def __init__(self, key):
        self.value = key
        self.left = None
        self.right = None

def zigzag(root):
    ans = []
    if root == None:
        return []    
    current_level = []
    next_level = []

    current_level.append(root)
    left_2_right = True

    while len(current_level) > 0:
        temp = current_level.pop(-1)
        ans.append(temp.value)
        if left_2_right:
            if temp.left != None:
                next_level.append(temp.left)
            if temp.right != None:
                next_level.append(temp.right)
        else:
            if temp.right != None:
                next_level.append(temp.right)
            if temp.left != None:
                next_level.append(temp.left)

        if len(current_level) == 0:
            left_2_right = not left_2_right
            current_level, next_level = next_level, current_level
    return ans

    

if __name__ == '__main__':
    # TREE 
    ####################################
    # root1 = Node(1)
    # root1.left = Node(2)
    # root1.right = Node(3)
    # root1.left.left = Node(4)
    # root1.left.right = Node(5)
    
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    root.right.left.right = Node(8)
    root.right.right.right = Node(9)
    root.right.right.right.right = Node(10)

    ####################################
    print(zigzag(root))
```

## Boundary Traversal of binary tree

```python
class Node:
    def __init__(self,key):
        self.value = key
        self.left = None
        self.right = None

def is_it_leaf_node(root):
    if root.left == None and root.right == None:
        return True
    return False

def left_traversal(root,ans):
    root = root.left
    while(root is not None):
        if not (is_it_leaf_node(root)):
            ans.append(root.value)
        if(root.left != None):
            root = root.left
        else:
            root = root.right

def right_traversal(root,ans):
    root = root.right
    stk = []
    while root is not None:
        if not (is_it_leaf_node(root)):
            stk.append(root.value)
        if(root.right != None):
            root = root.right
        else:
            root = root.left
    while(len(stk) != 0):
        ans.append(stk.pop(-1))

def add_leaf_node(root,ans):
    # pro order traversal
    if root is None:
        return
    if is_it_leaf_node(root):
        ans.append(root.value)
    add_leaf_node(root.left,ans)
    add_leaf_node(root.right,ans)


def Boundary_Traversal_BT(root):
    # logic goes from here...
    ans = []
    if root is None:
        return ans  
    if not (is_it_leaf_node(root)):
        ans.append(root.value)
    
    left_traversal(root,ans)
    add_leaf_node(root,ans)
    right_traversal(root,ans)
    
    return ans


if __name__ == '__main__':
    # TREE 
    ####################################
    # root1 = Node(1)
    # root1.left = Node(2)
    # root1.right = Node(3)
    # root1.left.left = Node(4)
    # root1.left.right = Node(5)
    
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.left = Node(6)
    root.right.right = Node(7)
    root.right.left.right = Node(8)
    root.right.right.right = Node(9)
    root.right.right.right.right = Node(10)

    ####################################
    ans = Boundary_Traversal_BT(root)
    print(ans)
```

## Maximum Path Sum in a Binary Tree

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None

def maxSumPath(root):
    if root is None:
        return 0
    
    l = maxSumPath(root.left)
    r = maxSumPath(root.right)

    max_at_node = max(max(l,r)+root.value, root.value)
    max_top = max(max_at_node , l+r+root.value)
    maxSumPath.ans = max(maxSumPath.ans, max_top)

    return max_at_node


def max_path_sum(root):
    maxSumPath.ans = float("-inf")
    maxSumPath(root)
    return maxSumPath.ans

if __name__ == '__main__':
    # TREE 
    ####################################
    # root = Node(1)
    # root.left = Node(2)
    # root.right = Node(3)
    # root.left.left = Node(4)
    # root.left.right = Node(5)
    # root.right.left = Node(6)
    # root.right.right = Node(7)
    # root.right.left.right = Node(8)
    # root.right.right.right = Node(9)
    # root.right.right.right.right = Node(10)

    root = Node(10)
    root.left = Node(2)
    root.right = Node(10)
    root.left.left = Node(20)
    root.left.right = Node(1)
    root.right.right = Node(-25)
    root.right.right.left = Node(3)
    root.right.right.right = Node(4)

    ####################################
    ans = max_path_sum(root)
    print(ans)
```

## Symmetric Tree (Mirror Image of itself)

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None

def isMirror(root1,root2):
    if root1 == None and root2 == None:
        return True

    if (root1 != None and root2 != None ) and root1.value == root2.value:
        if isMirror(root1.left, root2.right) and \
            isMirror(root1.right,root2.left):
            return True
    return False

def is_symmetric(root):
    return isMirror(root,root)

if __name__ == '__main__':
    root = Node(1)
    root.left = Node(2)
    root.right = Node(2)
    root.left.left = Node(3)
    root.left.right = Node(4)
    root.right.left = Node(4)
    root.right.right = Node(3)

    # root = Node(1)
    # root.left = Node(2)
    # root.right = Node(2)
    # root.left.left = Node(3)
    # root.left.right = Node(4)
    # root.right.left = Node(4)
    # root.right.right = Node(3)
    # root.right.left.right = Node(8)
    # root.right.right.right = Node(9)
    # root.right.right.right.right = Node(10)

    print(is_symmetric(root))
```

## Flatten a binary tree into linked list

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None

def flatten(root):
    while root != None:
        if root.left != None:
            # this pointer will go till the right most element of left child
            curr  = root.left
            while(curr.right != None):
                curr = curr.right
            # move right side of root to that right most node's child
            curr.right = root.right
            root.right = root.left
            root.left = None
            # update root (go one step ahead)
        root = root.right

def inorder(root):
    if root is None:
        return 
    inorder(root.left)
    print(root.value, end =" ")
    inorder(root.right)
    return
if __name__ == '__main__':
    root = Node(1)
    root.left = Node(2)
    root.right = Node(5)
    root.left.left = Node(3)
    root.left.right = Node(4)
    root.right.left = Node(6)
    root.right.right = Node(7)
    
    flatten(root)
    inorder(root)
```

## Check for Children Sum Property in a Binary Tree

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None

def is_this_children_sum_tree(root):
    left_data = 0
    right_data = 0
    if root is None or (root.left is None and root.right is None):
        return True
    else:
        if root.left != None:
            left_data = root.left.value
        
        if root.right != None:
            right_data = root.right.value

        if root.value == (left_data + right_data) and is_this_children_sum_tree(root.left) and is_this_children_sum_tree(root.right):
            return True
        return False

if __name__ == '__main__':
    root = Node(10)
    root.left = Node(8)
    root.right = Node(2)
    root.left.left = Node(3)
    root.left.right = Node(5)
    root.right.right = Node(2)

    if is_this_children_sum_tree(root):
        print("Tree is proper")
    else:
        print("Tree is not proper")
```

# Populate Next Right pointers of Tree

```python
# simple level order traversal and map left node with right node of tree......
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right= None
        self.next = None

def populate_next_right_pointer(root):
    if root is None:
        return None
    q = []
    q.append(root)
    child_q= []
    while q:        
        for node in q:
            if node.left != None:
                child_q.append(node.left)
            if node.right != None:
                child_q.append(node.right)

        for i in range(len(child_q)-1):
            child_q[i].next = child_q[i+1]

        q = child_q
        child_q = []
    return root

def print_next(root):
    if root is None:
        return root

    q = []
    q.append(root)
    while q:
        current = q.pop()
        check_none = lambda x: x.value if x is not None else None
        print(f"{current.value} ---> {check_none(current.next)}")
        if current.left != None:
            q.append(current.left)
        if current.right != None:
            q.append(current.right)


if __name__ == '__main__':
    root = Node(1)
    root.left = Node(2)
    root.left.left = Node(4)
    root.left.right = Node(5)

    root.left.left.left = Node(8)
    root.left.left.right = Node(9)

    root.left.right.left = Node(10)
    root.left.right.right = Node(11)

    root.right = Node(3)
    root.right.left = Node(6)
    root.right.right = Node(7)
    root.right.left.right = Node(13)
    root.right.right.left = Node(14)

    root = populate_next_right_pointer(root)

    print(root.right.next)
    print_next(root)
```

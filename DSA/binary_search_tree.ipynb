# Insertion in Binary Search Tree (BST)


```python
class Node:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None

def insert(root, value):
    if root is None:
        return Node(value)
    else:
        if value < root.value:
            root.left = insert(root.left, value)
        elif value > root.value:
            root.right = insert(root.right, value)
        else:
            # root.value == value
            return root
    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.value, end=" ")
        inorder(root.right)

if __name__ == '__main__':
    # Let us create the following BST
    # 50
    #  /     \
    # 30     70
    #  / \ / \
    # 20 40 60 80
 
    root = Node(50)
    # return value does not make any sence as it will return root
    # It will make sence when we will learn about 'balance Binary Search Tree'
    root = insert(root, 30)
    # print(root.value)
    root = insert(root, 20)
    # print(root.value)
    root = insert(root, 40)
    # print(root.value)
    root = insert(root, 70)
    # print(root.value)
    root = insert(root, 60)
    # print(root.value)
    root = insert(root, 80)
    # print(root.value)
    # ....................
    print("Inorder traversal (it must be sorted) : ",end=" ")
    inorder(root)
```

# Searching in Binary Search Tree (BST)

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None

def insert(root, value):
    if root is None:
        return Node(value)
    else:
        if value < root.value:
            root.left = insert(root.left, value)
        elif value > root.value:
            root.right = insert(root.right, value)
        else:
            # root.value == value
            return root
    return root

def search_element(root,value):
    if root is None or root.value == value:
        return root
    
    if root.value < value:
        return search_element(root.right, value)
    else:
        return search_element(root.left, value)

def inorder(root):
    if root:
        inorder(root.left)
        print(root.value, end=" ")
        inorder(root.right)

if __name__ == '__main__':
    # Let us create the following BST
    # 50
    #  /     \
    # 30     70
    #  / \ / \
    # 20 40 60 80
 
    root = Node(50)
    # return value does not make any sence as it will return root
    # It will make sence when we will learn about 'balance Binary Search Tree'
    root = insert(root, 30)
    # print(root.value)
    root = insert(root, 20)
    # print(root.value)
    root = insert(root, 40)
    # print(root.value)
    root = insert(root, 70)
    # print(root.value)
    root = insert(root, 60)
    # print(root.value)
    root = insert(root, 80)
    # print(root.value)
    # ....................
    key = [6,60]
    for i in key:
        if search_element(root,i) is None:
            print(f"{i} --> element is not there")
        else:
            print(f"{i} --> element is there")
```

# Deletion in Binary Search Tree (BST)
- intution is bit difficult

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.right = None
        self.left = None

def insert(root, value):
    if root is None:
        return Node(value)
    
    if root.value < value:
        root.right = insert(root.right, value)
    else:
        root.left = insert(root.left, value)
    return root

def inorder(root):
    if root == None:
        return
    inorder(root.left)
    print(root.value, end=" ")
    inorder(root.right)


def delete_element(root,value):
    if root is None:
        return root
    
    # travers till the value
    if root.value < value:
        root.right = delete_element(root.right, value)
        return root
    elif root.value > value:
        root.left = delete_element(root.left, value)
        return  root

    # Now root is equal to target value
    # apply all 3 conditions.....
    if root.left is None:
        temp = root.right
        del root
        return temp
    elif root.right is None:
        temp = root.left
        del root
        return temp
    else:
        # if both child are there
        current = root
        # find left most node on this child
        sub_root = root.right 
        while sub_root.left is not None:
            current = sub_root
            sub_root = sub_root.left
            
        if current != root:
           current.left = sub_root.right
        else:
            current.right = sub_root.right
        root.value = sub_root.value
        del sub_root
        return root


if __name__ == '__main__':
    root = None
    root = insert(root, 50)
    root = insert(root, 30)
    root = insert(root, 20)
    root = insert(root, 40)
    root = insert(root, 70)
    root = insert(root, 60)
    root = insert(root, 80)
    # root= insert(root, 76)
    # root= insert(root, 77)

    key = [50]
    for i in key:
        delete_element(root,i)
        inorder(root)
```

# Binary Search Tree (BST) Traversals â€“ Inorder, Preorder, Post Order

```python
# It is same as Binary Tree :)
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
        
def insert(root,value):
    if root is None:
        return Node(value)
    
    if root.value < value:
        root.right = insert(root.right, value)
    else:
        root.left = insert(root.left, value)
    return root

def inorder(root):
    if root == None:
        return
    inorder(root.left)
    print(root.value, end=" ")
    inorder(root.right)

def preorder(root):
    if root:
        print(root.value,end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.value,end=" ")


if __name__ == '__main__':
    root = None
    root = insert(root, 50)
    root = insert(root, 30)
    root = insert(root, 20)
    root = insert(root, 40)
    root = insert(root, 70)
    root = insert(root, 60)
    root = insert(root, 80)

    inorder(root)
    print("\n")
    preorder(root)
    print("\n")
    postorder(root)
```

# Convert a normal BST to Balanced BST

```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
        

def create_balanced_BST_from_sorted_arr(arr,start,end):
    if start > end:
        return None
    mid = (start + end)//2
    root = Node(arr[mid])

    root.left = create_balanced_BST_from_sorted_arr(arr,start,mid-1)
    root.right = create_balanced_BST_from_sorted_arr(arr,mid+1,end)

    return root

def create_pre_order_array(root,arr):
    if root:
        create_pre_order_array(root.left,arr)
        arr.append(root.value)
        create_pre_order_array(root.right,arr)


def make_all_things_work_to_gether(root):
    # creat sorted array and using that array creat balanced BST
    arr = []
    create_pre_order_array(root,arr)
    total = len(arr)
    new_root = create_balanced_BST_from_sorted_arr(arr,0,total-1)
    return new_root

def preorder(root):
    if root:
        print(root.value,end=" ")
        preorder(root.left)
        preorder(root.right)

if __name__ == '__main__':
    root = Node(10)
    root.left = Node(8)
    root.left.left = Node(7)
    root.left.left.left = Node(6)
    root.left.left.left.left = Node(5)
    print("before : ",end="")
    preorder(root)
    new_BST_root = make_all_things_work_to_gether(root)
    print("\nafter  : ",end="")
    preorder(new_BST_root)
```
